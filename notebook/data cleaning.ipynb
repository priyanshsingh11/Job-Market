```python
import pandas as pd
```

```python
df=pd.read_csv('raw/jobs_big_dataset.csv')
```

```python
df
```

```python
df.info()
```

```python
df.isna().sum()
df.describe(include='all')
```

```python
df=df.drop(columns=['salary_currency'])
```

```python
df.info()
```

```python
text_cols = ['title','company','location','description']
for col in text_cols:
    df[col] = df[col].astype(str).str.strip()
```

```python
df
```

```python
import re
df['description'] = df['description'].apply(lambda x: re.sub('<.*?>', '', str(x)))
```

```python
df
```

```python
# Calculate medians
median_min = df['salary_min'].median()
median_max = df['salary_max'].median()

# Fill NaN values
df['salary_min'] = df['salary_min'].fillna(median_min)
df['salary_max'] = df['salary_max'].fillna(median_max)
```

```python
df[['salary_min', 'salary_max']].isna().sum()
```

```python
df['salary_avg'] = (df['salary_min'] + df['salary_max']) / 2
```

```python
df
```

```python
df['city'] = df['location'].apply(lambda x: str(x).split(',')[0])
```

```python
df['employment_type'] = df['employment_type'].str.lower().str.replace('–','-')
```

```python
df
```

```python
df = df.dropna(subset=['title', 'company', 'description'])
```

```python
df
```

```python
df=df.drop(columns=['location','salary_min','salary_max'])
```

```python
df
```

```python
df['job_numeric_id'] = range(1, len(df) + 1)
```

```python
df
```

```python
df["posted_date"] = pd.to_datetime(df["posted_date"], errors="coerce")
df["posted_date"] = df["posted_date"].dt.date
```

```python
import numpy as np

# Step 1: Calculate IQR boundaries
Q1 = df["salary_avg"].quantile(0.25)
Q3 = df["salary_avg"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

median_salary = df["salary_avg"].median()

# Step 2: Replace outliers
df["salary_avg"] = np.where(
    (df["salary_avg"] < lower_bound) | (df["salary_avg"] > upper_bound),
    median_salary,
    df["salary_avg"]
)
```

```python
outliers = df[(df["salary_avg"] < lower_bound) | (df["salary_avg"] > upper_bound)]
print("Number of outliers replaced:", len(outliers))
```

```python
plt.figure(figsize=(10,16))
plt.hist(df["salary_avg"], bins=40)
plt.title("Salary Distribution After Fixing Outliers")
plt.xlabel("Salary Avg")
plt.ylabel("Frequency")
plt.show()
```
