```python
df["title"] = df["title"].str.replace(r"\d{4}", "", regex=True)
df["title"] = df["title"].str.replace(r"[-(),]", " ", regex=True)
df["title"] = df["title"].str.lower().str.strip()
```

```python
def extract_seniority(title):
    if "intern" in title:
        return "intern"
    if "junior" in title or "jr" in title:
        return "junior"
    if "senior" in title or "sr" in title:
        return "senior"
    if "lead" in title:
        return "lead"
    if "manager" in title:
        return "manager"
    if "principal" in title:
        return "principal"
    return "mid"

df["seniority_level"] = df["title"].apply(extract_seniority)
```
